```python
import keras
```

```python
train_data = []
with open("massaged.msv") as m:
    for line in m:
        train_data.append(list(map(float, line.split(','))))
```

```python
train_data[0]
```

```python
train_labels = []
with open("survived.msv") as s:
    for line in s:
        train_labels.append(int(line))
```

```python
train_labels[0]
```

```python
import numpy as np
```

```python
y_train = np.asarray(train_labels).astype('float32')
```

```python
x_train = np.asarray(train_data).astype('float32')
```

```python
x_train[0]
```

```python
from keras import models
from keras import layers

model = models.Sequential()
model.add(layers.Dense(16, activation="relu", input_shape=(6,)))
model.add(layers.Dense(16, activation="relu"))
```

```python
model.add(layers.Dense(1, activation="sigmoid"))
```

```python
model.compile(optimizer='rmsprop', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
x_val = x_train[:400]
partial_x_train = x_train[400:]
y_val = y_train[:400]
partial_y_train = y_train[400:]
```

```python
history = model.fit(partial_x_train, partial_y_train, epochs=20, batch_size=16, validation_data=(x_val, y_val))
```

```python
test_data = []
with open("testm.msv") as m:
    for line in m:
        test_data.append(list(map(float, line.split(','))))
```

```python
x_test = np.asarray(test_data).astype('float32')
```

```python
p = model.predict(x_test)
```

```python
ans = np.around(p)
```
